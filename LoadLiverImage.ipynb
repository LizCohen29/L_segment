```
def LoadLiverImage():
    Train_LIB = '/content/gdrive/Team Drives/Liver_Project/TrainData/ct/train/'
    Val_LIB = '/content/gdrive/Team Drives/Liver_Project/TrainData/ct/val/'
    MASKTrain_LIB = '/content/gdrive/Team Drives/Liver_Project/TrainData/seg/train/'
    MASKVal_LIB = '/content/gdrive/Team Drives/Liver_Project/TrainData/seg/val/'

    IMG_HEIGHT, IMG_WIDTH = 512, 512
    SEED=42


    x_train = [x for x in sorted(os.listdir(Train_LIB)) if os.path.isfile(os.path.join(Train_LIB,x))] # Read train images
    x_val = [x for x in sorted(os.listdir(Val_LIB)) if os.path.isfile(os.path.join(Val_LIB,x))] # Read validation images
    y_train = [x for x in sorted(os.listdir(MASKTrain_LIB)) if os.path.isfile(os.path.join(MASKTrain_LIB,x))] # Read train segmentation
    y_val = [x for x in sorted(os.listdir(MASKVal_LIB)) if os.path.isfile(os.path.join(MASKVal_LIB,x))] # Read validation segmentation
    X_train = np.empty((len(x_train), IMG_HEIGHT, IMG_WIDTH), dtype='float32')

    for i, name in enumerate(x_train):
        im = cv2.imread(Train_LIB + name, cv2.IMREAD_UNCHANGED).astype('int16').astype('float32')
        im = (im - np.min(im)) / (np.max(im) - np.min(im)) # Normalization between 0-1
        X_train[i] = im[:,:,1]
        if np.mod(i,100)==0:
          print(i)

    Y_train = np.empty((len(y_train), IMG_HEIGHT, IMG_WIDTH), dtype='float32')
    for i, name in enumerate(y_train):
        im = cv2.imread(MASKTrain_LIB + name, cv2.IMREAD_UNCHANGED).astype('float32')/255.
        Y_train[i] = im[:,:,1]
        if np.mod(i,100)==0:
          print(i)

    fig, ax = plt.subplots(1,2, figsize = (8,4))
    plt.axis('off')
    ax[0].imshow(X_train[0], cmap='gray')
    ax[0].grid(False)
    ax[1].imshow(Y_train[0], cmap='gray')
    ax[1].grid(False)
    plt.show()


    X_val = np.empty((len(x_val), IMG_HEIGHT, IMG_WIDTH), dtype='float32')
    for i, name in enumerate(x_val):
        im = cv2.imread(Val_LIB + name, cv2.IMREAD_UNCHANGED).astype('int16').astype('float32')
        im = (im - np.min(im)) / (np.max(im) - np.min(im)) # Normalization between 0-1
        X_val[i] = im[:,:,1]
        if np.mod(i,100)==0:
          print(i)

    Y_val = np.empty((len(y_val), IMG_HEIGHT, IMG_WIDTH), dtype='float32')
    for i, name in enumerate(y_val):
        im = cv2.imread(MASKVal_LIB + name, cv2.IMREAD_UNCHANGED).astype('float32')/255.
        Y_val[i] = im[:,:,1]
        if np.mod(i,100)==0:
          print(i)
    fig, ax = plt.subplots(1,2, figsize = (8,4))
    plt.axis('off')
    ax[0].imshow(X_val[0], cmap='gray')
    ax[0].grid(False)
    ax[1].imshow(Y_val[0], cmap='gray')
    ax[1].grid(False)
    plt.show()    

    # Add channel dimension
    X_train = X_train[:,:,:,np.newaxis] 
    X_val = X_val[:,:,:,np.newaxis]
    Y_train = Y_train[:,:,:,np.newaxis]
    Y_val = Y_val[:,:,:,np.newaxis]
    
return  X_train, X_val, Y_train, Y_val
```
